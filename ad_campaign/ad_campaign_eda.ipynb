```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ad_campaign_data.csv')
df.head()
```

```python
print(df.shape, df.columns, df.describe(), sep = "\n\n")
```

```python
numerical_variables = df.select_dtypes(include=["number"]).columns.tolist()
```

```python
for var in numerical_variables:
    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    sns.histplot(df[var], kde=True, ax=ax[0])
    sns.boxplot(x=df[var], ax=ax[1])
    plt.suptitle(f'Distribution of {var}')
    plt.tight_layout()
    plt.show()
```
