```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset_Ads.csv')
df.head()
```

```python
df['Click Time'] = pd.to_datetime(df['Click Time'])
```

```python
df = df.query('(Age >= 18) & (Age <= 65) & (Income > 0)').copy()
```

```python
df.shape
```

```python
numerical_variables = list(df.select_dtypes(include="number").columns)
```

```python
for var in numerical_variables:
    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    sns.histplot(df[var], kde=True, ax=ax[0])
    sns.boxplot(x=df[var], ax=ax[1])
    plt.suptitle(f'Distribution of {var}')
    plt.tight_layout()
    plt.show()
```
